Import the tools I need.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import json
```

Link to my National Park Service api key

```python
with open('../data/cap_keys.json') as fi:
    credentials = json.load(fi)
```

```python
api_key = credentials['api_key_nps']
```

## New path - looking at passport stamp locations

```python
endpoint = "https://developer.nps.gov/api/v1/passportstamplocations/"

HEADERS = {"X-Api-Key":api_key}

stamps = requests.get(endpoint,headers=HEADERS)
```

```python
stamps
```

```python
stamps = stamps.json()
```

```python
stamps
```

This tells me there are 1001 places but it is only pulling in 50 at a time.
